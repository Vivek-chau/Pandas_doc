### How to handle time series data with ease using pandas datetime properties.

```python
# I want to work with the dates in the column datetime as datetime objects instead of plain text

air_quality_no2['datetime'] = pd.to_datetime(air_quality_no2["datetime"])
```

```python
air_quality_no2['datetime'].dtypes
```

#### Initially, the values in datetime are character strings and do not provide any datetime operations (e.g. extract the year, day of the week,…).
#### By applying the to_datetime function, pandas interprets the strings and convert these to datetime (i.e. datetime64[ns, UTC]) objects.

```python
# What is the start and end date of the time series data we are working with?
```

```python
air_quality_no2['datetime'].min(), air_quality_no2['datetime'].max()
```

```python
# Now we can use this to get the length of our time series.
```

```python
air_quality_no2['datetime'].max()-air_quality_no2['datetime'].min()
```

```python
# Now i want to add a new column to the DataFrame containing only the month of the measurement.
```

```python
air_quality_no2['month']= air_quality_no2['datetime'].dt.month
```

```python
air_quality_no2.head()
```

```python
# What is the average NO2 concentration for each day of the week for each of the measurement locations?
# To group on weekdays, we use the datetime property weekday (with Monday=0 and Sunday=6) of pandas Timestamp, which is also accessible by the dt accessor. 
# The grouping on both locations and weekdays can be done to split the calculation of the mean on each of these combinations.

```

```python
air_quality_no2.groupby([air_quality_no2['datetime'].dt.weekday, 'location'])['value'].mean()
```

```python
# Ques -What is the avg value for each hour of the day?
fig, axs = plt.subplots(figsize=(12,4))
air_quality_no2.groupby(air_quality_no2['datetime'].dt.hour)['value'].mean().plot(kind='bar',rot=0, ax=axs)
plt.xlabel("Hour of the day")
plt.ylabel("$NO_2 (µg/m^3)$")
plt.show()
```

#### Resample a time series to another frequency

```python
# Aggregate the current hourly time series values to the monthly maximum value in each of the stations.
```

```python
# This will resample your time series data to the end of each month and then calculate the maximum value for each resampled period.
monthly_max = no_2.resample('M').max()
monthly_max
```

```python
monthly_max.index.freq
```

```python
no_2.resample("D").mean().plot(style="-o", figsize=(10,5))
plt.show()
```
